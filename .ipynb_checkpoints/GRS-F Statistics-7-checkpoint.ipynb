```python
import numpy as np
import pandas as pd
from pathlib import Path
import os
import statsmodels.api as sm
import pickle
```

```python
lis = pd.DataFrame(os.listdir("/home/harshit/Project/complete_models/"))
coeff = pd.DataFrame(columns = ["Symbol","const","SMB %", "HML %","Rm-Rf %", "high-low"] )
#Making a dataframe of all coffecients
for i in lis[0]:
    data = open("/home/harshit/Project/complete_models/"+i,"rb")
    symbol = i.split('.')[0]
    model = pickle.load(data)
    frame = pd.DataFrame(model.params).T
    frame.insert(0,"Symbol",symbol)
    coeff = coeff.append(frame,ignore_index=True,sort=False)
intercept =coeff['const']
```

```python
all_factors= pd.read_csv("/home/harshit/Project/allfactors.csv")
```

```python
#Calculating residuals
model_path = "/home/harshit/Project/complete_models/"
return_path = "/home/harshit/Project/Returns/all_ret_by_company/"
mod = os.listdir(model_path)
ret = os.listdir(return_path)
residuals = pd.DataFrame()
for i in ret:
    i = i.split(".")[0]
    equity = pd.read_csv(return_path+i+".csv")                       # loading daily Return data
    equity.Date = equity.Date.apply(lambda x: x.replace("-",""))
    equity["Date"] = equity["Date"].apply(pd.to_numeric)
    equity=equity.drop([0])
    
    all_data = pd.merge(all_factors,equity,on="Date")
    all_data.rename(columns={'Return':i}, inplace=True)
    X= all_data[["SMB %","HML %","WML %","Rm-Rf %","high-low"]] #Making X variable for entropy included model
    X= sm.add_constant(X)
    y = all_data[i]
    
    
    data = open("/home/harshit/Project/complete_models/"+i+".sav","rb") # loading model
    symbol = i.split('.')[0]
    model = pickle.load(data)
    residual = y - pd.DataFrame(model.predict(X),columns=[i])
    residual= pd.merge(all_data[["Date"]],residual,left_index=True, right_index=True)
    residual = residual.T
    residuals = residuals.append(residual)
residuals.reset_index(inplace=True)
```
